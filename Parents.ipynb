```python
with open('no_pii_grievance_v2\\no_pii_grievance_v2.json', encoding="utf8") as json_file:
    data = json.load(json_file)
```

```python
extracted_data = []
for i in range(35):
    for k,v in data[i].items():
        if "v7_target" in k:
            if(data[i]["v7_target"]=="Yes" and data[i]["_id"]!="DEPOJ/E/2023/0000002"):
                extracted_data.append({
                    "_id" : data[i]["_id"],
                    "code" : data[i]["CategoryV7"]["$numberLong"],
                    "subject_content_text" : data[i]["subject_content_text"]
                })

output_file = "extracted_data.json"
with open(output_file, "w") as f:
    for item in extracted_data:
        json.dump(item, f)
        f.write('\n')

print("Extracted data has been written to", output_file)
```
