```python
import json
```

```python
with open('output.json', encoding="utf8") as json_file:
    data_2 = json.load(json_file)
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('CategoryCode_Mapping_v2.csv')
```

```python
df['Parent'] = df['Parent'].fillna(-1).astype(int)
```

```python
df.head()
```

```python
for i in range(len(data_2)):
    par = int(data_2[i]['code'])
    p = df[df['Code']==par]['Parent']
    if not p.empty:
        while p.values[0]!=-1:
            print(par)
            par = int(p.values[0])
            p = df[df['Code'] == par]['Parent']
            print(par)
            # print(type(p.values[0]))
        s = df[df['Code'] == par]['Description'].values[0]
        print(type(s))
        data_2[i]["parent_name"] = s
```

```python
dk = pd.DataFrame(data_2)
dk.head()
# dk.drop(columns=['_id', 'code'], inplace=True)
```

```python
empty_parent_rows = dk[dk['parent_name'].isnull()]

# Save the selected columns of the filtered rows to a new CSV file
empty_parent_rows[['_id', 'code', 'summarized_text']].to_csv('test.csv', index=False)
```

```python
dk.drop(columns=['_id', 'code'], inplace=True)
```

```python
empty_parent_rows = dk[dk['parent_name'].notnull()]

# Save the selected columns of the filtered rows to a new CSV file
empty_parent_rows[['summarized_text', 'parent_name']].to_csv('train.csv', index=False)
```

```python
output_file_csv = "data_final.csv"
dk.to_csv(output_file_csv, index=False)

print("Extracted data has been written to", output_file_csv)
```
